```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,10,20,30,30,30,40,60,70,75]
weight_list = [20,25,60,85,86,87,70,90,95,85]
```

```python
plt.plot(age_list, weight_list, "r")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("Age vs Weight")
plt.show()
```

```python
plt.savefig("test.png")
```

```python
np_age = np.array(age_list)
np_weight = np.array(weight_list)
```

```python
plt.plot(np_age, np_weight, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
np_arr1 = np.linspace(0,10,20)
```

```python
np_arr1
```

```python
np_arr2 = np_arr1 ** 3
```

```python
np_arr2
```

```python
plt.plot(np_arr1, np_arr2)
plt.show()
```

```python
plt.plot(np_arr1, np_arr2, "r*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(np_arr1, np_arr2, "r*-")
plt.subplot(1,2,2)
plt.plot(np_arr2, np_arr1, "b--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.8,0.8]) #Scale of figure.
figure_axes.plot(np_arr1, np_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=125)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(np_arr1, np_arr1 ** 2, label="numpy array **2")
new_axes.plot(np_arr1, np_arr1 ** 3, label="numpy array **3")
new_axes.legend(loc=0)
plt.show()
```
