```python
df = pd.read_csv("8-apply_function_data.csv")
```

```python
def categorize_salaries(salary:int) -> str:
    if salary < 50000:
        return "Low"

    if 50000 <= salary < 80000:
        return "Medium"

    return "High"
```

```python
df["Salary_Category"] = df["Salary"].apply(categorize_salaries)
```

```python
df.head()
```

```python
def categorize_performances(entry):
    performance_score = entry["Performance_Score"]
    experience = entry["Experience"]
    
    if (experience >= 10):
        performance_score+=1

    return performance_score
```

```python
df["New_Performance_Score"] = df.apply(categorize_performances, axis=1)
```
