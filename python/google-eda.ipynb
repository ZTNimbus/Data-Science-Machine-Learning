```python
chars_to_remove = ["+", ",", "$"]
cols_to_clean = ["Installs", "Price"]
```

```python
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_clean[cols] = df_clean[cols].str.replace(item, "")
```

```python
df_clean.tail()
```

```python
df_clean["Price"].unique()
```

```python
df_clean["Installs"].unique()
```

```python
df_clean["Price"] = df_clean["Price"].astype(float)
df_clean["Installs"] = df_clean["Installs"].astype(int)
```

```python
df_clean.describe()
```

```python
df_clean["Last Updated"]
```

```python
df_clean["Last Updated"] = pd.to_datetime(df_clean["Last Updated"])
```

```python
df_clean.head()
```

```python
df_clean.info()
```

```python
df_clean["Day"] = df_clean["Last Updated"].dt.day
df_clean["Month"] = df_clean["Last Updated"].dt.month
df_clean["Year"] = df_clean["Last Updated"].dt.year
```

```python
df_clean.head()
```

```python
df_clean[df_clean.duplicated("App")].shape
```

```python
df_clean = df_clean.drop_duplicates(subset=["App"], keep="first")
```

```python
df_clean.shape
```

```python
numeric_features = [feature for feature in df_clean.columns if df_clean[feature].dtype != "O"]
categorical_features = [feature for feature in df_clean.columns if df_clean[feature].dtype == "O"]
```

```python
print(numeric_features)
```

```python
plt.figure(figsize=(15,10))

for i in range(0, len(numeric_features)):
    plt.subplot(5,3, i+1)
    sns.kdeplot(x=df_clean[numeric_features[i]], color="b", fill=True)
    plt.xlabel(numeric_features[i])
    plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15,6))

categories = ["Type", "Content Rating"]

for i in range(0, len(categories)):
    plt.subplot(1, 2, i+1)
    sns.countplot(x=df_clean[categories[i]], color="b", fill=True)
    plt.xlabel(categories[i])
    plt.tight_layout()

plt.show()
```

```python
df_category_installs = df_clean.groupby(["Category"])["Installs"].sum().sort_values(ascending=False).reset_index()
```

```python
df_category_installs
```

```python
df_category_installs["Installs"] = df_category_installs["Installs"] / 1_000_000_000
```

```python
df_category_installs
```

```python
df2 = df_category_installs.head(10)
```

```python
df2
```

```python
plt.figure(figsize=(15,6))

sns.barplot(x="Installs", y="Category", data=df2)
plt.xlabel("Installs(In Billions)")
plt.title("Top 10 Categories by Installs")

plt.show()
```

```python
df_app_cat = df_clean.groupby(["Category", "App"])["Installs"].sum().reset_index()
```

```python
df_app_cat
```

```python
df_app_cat = df_app_cat.sort_values("Installs", ascending=False)
```

```python
df_app_cat
```

```python
apps = ["GAME", "COMMUNICATION", "TOOLS", "PRODUCTIVITY", "SOCIAL"]
```

```python
plt.figure(figsize=(15,15))

for i, app in enumerate(apps):
    df2 = df_app_cat[df_app_cat["Category"] == app]
    df2 = df2.head(5)

    plt.subplot(3, 2, i+1)

    sns.barplot(x="Installs", y="App", data=df2)
    plt.title(app, size=20)

plt.tight_layout()

plt.show()
```
