```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.bar(df["Geography"], df["EstimatedSalary"].mean(), color="blue", edgecolor="orange")
plt.title("Geograpy VS Estimated Salary")
plt.show()
```

```python
plt.scatter(df["Age"], df["Tenure"])
plt.title("Age VS Tenure Scatter")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.hist(df["Tenure"], bins=30)
plt.ylabel("Probability")
plt.xlabel("Tenure")
plt.show()
```
