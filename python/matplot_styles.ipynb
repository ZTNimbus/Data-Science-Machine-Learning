```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data2 = data1 ** 2
```

```python
my_fig, my_axes = plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2)
my_axes.plot(data2, data1)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="#151515", linewidth=1.0)
new_axes.plot(data1, data1 + 3, color="#909090", linewidth=2.0)
new_axes.plot(data1, data1 + 4, color="#111", linestyle="-.")
new_axes.plot(data1, data1 + 5, color="#111", linestyle=":")
new_axes.plot(data1, data1 + 6, color="#111", linestyle="--", marker="o", markersize=8, markerfacecolor="red")
new_axes.plot(data1, data1 + 7, color="#111", linestyle="--", marker="+", markersize=10)
plt.show()
```

```python
plt.scatter(data1, data2)
plt.show()
```
