```python
X = df.drop("selling_price", axis=1)

y = df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=15, test_size=0.3)
```

```python
cat_cols = df.select_dtypes("object").columns.to_list()
```

```python
cat_cols
```

```python
df[cat_cols].nunique()
```

```python
# Onehot encoding for --> seller_type, transmission_type, fuel_type. Reasoning: Low number of unique values
# Frequency encoding for --> car_name, brand, model. Reasoning: High number of unique values + other encoding types do not meaningfully work
```

```python
one_hot_cols = ["seller_type", "transmission_type", "fuel_type"]
freq_cols = ["model", "brand", "car_name"]
```

```python
for col in freq_cols:
    freq = X_train[col].value_counts() / len(X_train)

    X_train[col + "_freq"] = X_train[col].map(freq)
    X_test[col + "_freq"] = X_test[col].map(freq)

    mean_freq = freq.mean()

    X_test[col + "_freq"] = X_test[col + "_freq"].fillna(mean_freq)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_train.drop(["car_name", "brand", "model"], axis=1, inplace=True)
X_test.drop(["car_name", "brand", "model"], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("one_hot", OneHotEncoder(drop="first", handle_unknown="ignore"), one_hot_cols)
    ],
    remainder="passthrough"
)
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
encoded_cols = preprocessor.get_feature_names_out()
```

```python
encoded_cols
```

```python
X_train = pd.DataFrame(X_train, columns=encoded_cols)
X_test = pd.DataFrame(X_test, columns=encoded_cols)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
```

```python
ada = AdaBoostRegressor()
```

```python
ada.fit(X_train, y_train)
y_pred = ada.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
#Hyperparameter tuning
```

```python
param_grid = {
    "n_estimators": [50, 80, 100, 120],
    "learning_rate": [0.001, 0.01, 0.1, 1.0, 2.0],
    "loss": ["linear", "square", "exponential"]
}
```

```python
rscv = RandomizedSearchCV(estimator=AdaBoostRegressor(), param_distributions=param_grid, scoring="r2", cv=5, n_jobs=-1)
```

```python
rscv.fit(X_train, y_train)
```

```python
rscv.best_params_
```

```python
y_pred = rscv.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
param_grid = {
    "estimator__max_depth": [3, 4, 5],
    "n_estimators": [50, 80, 100, 120],
    "learning_rate": [0.001, 0.01, 0.1, 1.0, 2.0],
    "loss": ["linear", "square", "exponential"]
}
```

```python
rscv = RandomizedSearchCV(estimator=AdaBoostRegressor(DecisionTreeRegressor()), param_distributions=param_grid, scoring="r2", cv=5, n_jobs=-1)
```
