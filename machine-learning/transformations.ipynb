```python
import math

def plot_all_histograms(df: pd.DataFrame, title_prefix: str = ""):

    num_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols) / n_cols)

    plt.figure(figsize=(5 * n_cols, 4 * n_rows))

    for i , col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f"{title_prefix} {col}")
        plt.xlabel("")
        plt.ylabel("")

    plt.tight_layout()
    plt.show()
```

```python
plot_all_histograms(df, "Original")
```

```python
X = df.drop("MEDV", axis=1)

y = df["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt_X = PowerTransformer(method="yeo-johnson")
```

```python
X_train_transformed = pt_X.fit_transform(X_train)
X_test_transformed = pt_X.transform(X_test)
```

```python
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=X_train.columns)
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=X_train.columns)
```

```python
plot_all_histograms(X_train_transformed_df, "Transformed")
```

```python
plot_all_histograms(df, "Original")
```

```python
from scipy.stats import boxcox
```

```python
y_train_transformed, lambda_y = boxcox(y_train)
```

```python
def inverse_boxcox(y, lambda_):
    if lambda_ == 0: return np.exp(y)
    return np.power(y * lambda_ + 1, 1 / lambda_)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train_transformed, y_train_transformed)
```

```python
y_pred_transformed = model.predict(X_test_transformed)
```

```python
y_pred_original = inverse_boxcox(y_pred_transformed, lambda_y)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print("R2:", r2_score(y_test, y_pred_original))
print("MAE:", mean_absolute_error(y_test, y_pred_original))
print("MSE:", mean_squared_error(y_test, y_pred_original))
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("R2:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
